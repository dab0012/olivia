```python


# Load librariesio data
path = '/home/dnllns/Documentos/ubu/TFG/libraries/chunks'

librariesio_df = pd.DataFrame()

# List all csv files in path
chunks = [f for f in os.listdir(path)]

for chunk in tqdm.tqdm(chunks):

    chunk_df = pd.read_csv(os.path.join(path, chunk), low_memory=True)

    # Drop columns ID,Project ID,Version ID,Dependency Platform,Dependency Kind,Optional Dependency,Dependency Requirements,Dependency Project ID
    chunk_df = chunk_df.drop(columns=['ID','Project ID','Version ID','Dependency Platform','Dependency Kind','Optional Dependency','Dependency Requirements','Dependency Project ID'])

    # Filter by package manager
    chunk_df = chunk_df[chunk_df['Platform'] == 'NPM']

    #append chunk to librariesio_df
    librariesio_df = pd.concat([librariesio_df, chunk_df])
```

```python
# Filter only for Python
librariesio_df.to_csv('/home/dnllns/Documentos/ubu/TFG/libraries/librariesio_npm.csv', index=False)
```

```python
package_version = {}

# Loop all the rows
for index, row in tqdm.tqdm(librariesio_df.iterrows(), total=librariesio_df.shape[0]):
    # Get the version number
    version_number = row['Version Number']
    # Get the project name
    project_name = row['Project Name']

    if project_name not in package_version:
        package_version[project_name] = version_number
    else:
        if version_number != package_version[project_name]:
            package_version[project_name] = version_number
```

```python
# Save the dictionary as a pickle file
import pickle


with open('/home/dnllns/Documentos/ubu/TFG/libraries/package_version_npm.pickle', 'wb') as handle:
    pickle.dump(package_version, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
```

```python
# Identify the rows to keep
mask = librariesio_df['Version Number'] == librariesio_df['Project Name'].map(package_version)

# Filter the dataframe
librariesio_df = librariesio_df[mask]
```
